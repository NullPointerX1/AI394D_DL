```python
import torch
import torch.distributions as dist
import matplotlib.pyplot as plt
```

```python
# Bernoulli distribution

bernoulli = dist.Bernoulli(torch.tensor([0.5]))

# Sample from the distribution
for _ in range(10):
    print(bernoulli.sample())
```

```python
samples = [bernoulli.sample() for _ in range(1000)]
samples = torch.stack(samples)

print(f"Mean: {torch.mean(samples)}")
print(f"Variance: {torch.var(samples)}")

plt.figure(figsize=(6,4))
plt.hist(samples.numpy(), bins=[-0.5,0.5,1.5], rwidth=0.6)
plt.xticks([0,1])
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Bernoulli Distribution Samples (n=1000)")
plt.show()
```

```python
# Normal distribution
normal = dist.Normal(torch.tensor([0.0]), torch.tensor([1.0]))
for _ in range(1000):
    print(normal.sample())
```

```python
samples = [normal.sample() for _ in range(1000)]
samples = torch.stack(samples)

print(f"Mean: {torch.mean(samples)}")
print(f"Variance: {torch.var(samples)}")

plt.figure(figsize=(7, 4))
plt.hist(samples.numpy(), bins=40, density=True)
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Normal Distribution Samples (n=1000)")
plt.grid(True)
plt.show()
```
